```python
list_of_files = [
    "Merged Employment/Resources/output_employment_data/Unified2017.csv",
    "Merged Employment/Resources/output_employment_data/Elementary2017.csv",
    "Merged Employment/Resources/output_employment_data/Secondary2017.csv"
]

# Format pd.read_csv(filename).assign(year=2017)
employment_2017_df = []
for i,filename in enumerate(list_of_files):
    X1 = pd.read_csv(filename)
    X1["year"] = 2017
    employment_2017_df.append(X1)
    
# Combine the data into a single dataset
employment_2017_data = pd.concat(employment_2017_df)
employment_2017_merged = pd.DataFrame(employment_2017_data)
```

```python
# Save csv file 

employment_2017_merged.to_csv("Merged Employment/Resources/merged_employment_yearly/Merged_Employment_2017.csv")
```
