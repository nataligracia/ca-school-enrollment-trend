```python
list_of_files = [
    "../Merged Employment/Resources/output_employment_data/Unified2019.csv",
    "../Merged Employment/Resources/output_employment_data/Elementary2019.csv",
    "../Merged Employment/Resources/output_employment_data/Secondary2019.csv"
]

# Format pd.read_csv(filename).assign(year=2019)
employment_2019_df = []
for i,filename in enumerate(list_of_files):
    X1 = pd.read_csv(filename)
    X1["year"] = 2019
    employment_2019_df.append(X1)
    
# Combine the data into a single dataset
employment_2019_data = pd.concat(employment_2019_df)
employment_2019_merged = pd.DataFrame(employment_2019_data)
```

```python
# Save csv file 

employment_2019_merged.to_csv("../Merged Employment/Resources/merged_employment_yearly/Merged_Employment_2019.csv")
```
