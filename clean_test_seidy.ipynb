```python
import pandas as pd
import pathlib
```

```python
school_directory_filepath = "Resources/ca-school-list.csv"
sy_1516_filpath = "Resources/2015-16-enrollment.csv"
```

```python
school_directory_df = pd.read_csv(school_directory_filepath)
school_directory_df = school_directory_df.rename(columns = {"CDSCode":"CDS_CODE"})
school_directory_df
```

```python
sy_1516_df = pd.read_csv(sy_1516_filpath)
sy_1516_df.head()
```

```python
enroll_test = sy_1516_df[["CDS_CODE", "COUNTY", "DISTRICT", "SCHOOL", "ENR_TOTAL"]]
total_enroll_test = enroll_test.groupby(by="CDS_CODE").sum()
```

```python
total_enrollment_series = total_enroll_test["ENR_TOTAL"]
```

```python
de_duped_test = enroll_test.drop_duplicates(subset="CDS_CODE")
re_indexed = de_duped_test.set_index("CDS_CODE")
re_indexed["Total Enrollment"] = total_enrollment_series
new_df = re_indexed[["COUNTY", "DISTRICT", "SCHOOL", "Total Enrollment"]]
new_df
```

```python
merged_df = pd.merge(new_df, school_directory_df, on="CDS_CODE", how="left")
merged_df
```

```python
merged_df.to_csv("merged_test.csv")
```
