```python
list_of_files = [
    "../Merged Employment/Resources/output_employment_data/Unified2016.csv",
    "../Merged Employment/Resources/output_employment_data/Elementary2016.csv",
    "../Merged Employment/Resources/output_employment_data/Secondary2016.csv"
]

# Format pd.read_csv(filename).assign(year=2016)
employment_2016_df = []
for i,filename in enumerate(list_of_files):
    X1 = pd.read_csv(filename)
    X1["year"] = 2016
    employment_2016_df.append(X1)
    
# Combine the data into a single dataset
employment_2016_data = pd.concat(employment_2016_df)
employment_2016_merged = pd.DataFrame(employment_2016_data)
```

```python
# Save csv file 

employment_2016_merged.to_csv("../Merged Employment/Resources/merged_employment_yearly/Merged_Employment_2016.csv")
```
